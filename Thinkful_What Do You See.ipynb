## Apple Stock: From IPO to January 22, 2019

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
from pandas import Series
import calendar
import datetime as dt
%pylab inline
%matplotlib inline
```

```python
# Loading the data into a data frame.
df = pd.read_csv('AAPL.csv')
df.head()
```

### Apple Stock Percent Change Over Time

```python
plt.plot(df['changePercent'])
plt.ylabel ('Percent Change')
plt.xlabel ('Trading Days Since IPO')
plt.title('GRAPH 1: Apple Stock Change Over Time')

# Changing the graph display size.
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16.5, 8.5)
plt.show()
```

<p>This graph shows the variance in percent change since Apple first realease its IPO in 2014. From this plot one can see that there were some days where the stock price varied greatly, though overall the price change each day seems to grow somewhat steadier as the years pass.</p>
<p></p>
<p>We can see that in its infancy, Apples stock was unsteady, with a percent decline greater than 7.5%. This decline was not surpassed until recently. The data shows that in 2019, there was a decline of almost 10%. This rapid swing can have many factors which might include its earnings call in January 2019. </p>

```python
plt.hist(df['change'], color='green')
plt.hist(df['changeOverTime'], color='blue')
plt.title('GRAPH 2: Percent Change Over Time')
plt.legend

# Changing the graph display size.
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16.5, 8.5)
plt.show()
```

<p>In the above histogram the color green is representative of the stocks <i>Change</i> and the blue is representative of <i>Change Over Time</i>.<p>
<p>From the right-skewed shape  of both <i>Change</i> and <i>Change Over Time</i> we can also infer that the mean is greater greater that the median. Though Apple has had swings in price, as seen from Graph 1, we can see from the histogram above that overall the stock price shows an increase over time.</p> 

### Stock Price Over Time

```python
df_2 = pd.read_csv('AAPL.csv')

# Changing 'date' from object to datetime
df_2['date'] = pd.to_datetime(df_2['date'])

# Creating line plot
df_2[['open','close','date']].plot(x='date')
plt.xlabel('Year')
plt.ylabel('Closing Amount in US $')
plt.title('GRAPH 3: Apple Open and Closing Stock Price Over Time')

# Changing the graph display size.
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16.5, 8.5)
plt.show()
```

<b>GRAPH 3</b> displays the open and close price of Apple stock since its IPO release in 2014. Overall the stock shows an upward trend with stock price climbing to over $225.00 in 2019.

The opening and closing price are relatively the same, with both lines being almost identical. Note the dip around mid-2105 in the opening price. Though the closing price eventually jumped back up to the meet the opening price of that short decline, it was the beginning of a downward trend in stock price. This trend persisted through the later part of 2015 and continued to all of 2016, finally gaining an upswing in 2017 that continued a steep increase until a sharp drop in in 2019.


```python
df_2[['changePercent','changeOverTime','date']].plot(x='date')
plt.ylabel ('Change')
plt.xlabel ('Year')
plt.title('GRAPH 4: Apple Stock Yearly Highs and Lows')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16.5, 8.5)
plt.show()
```

<b>GRAPH 4</b> displays the percent change in daily stock price and the percent change of stock over time. We can see that the stock price in a given time can vary greatly, dropping as low as -10% and increasing to +7% a short time later at the tail end of 2019. However, change over time is less erratic and appears to have a slight ascending trend. Taken all together, this is indicative of Apples' overall stability and possible status as a growth stock in the market.

### Trading Volume

```python
df_2[['volume','date']].plot(x='date')
plt.ylabel ('Trading Volumein Millions')
plt.xlabel ('Year')
plt.title('GRAPH 5: Trading Volume')
# Changing the graph display size.
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16.5, 8.5)
plt.show()
```

In <b>GRAPH 5</b> we can see the trading volume over the last five years. Stocks tend to trade at a very high volume on the first few days of releasing to the public. This same trend can be seen with Apple stock as the highest trading volume was seen during the beginning day of its IPO release. Trading volume never reaches the high of +2.5 million again after that time. Over time we can also see that trading volume is decreasing. If we compare this to Graph 3, it appears that higher trading occurs in tandem with an increase in stock price though further analysis is needed to confirm.
